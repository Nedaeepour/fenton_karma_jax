```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import io
```

```python
def plot(results):
    results = pd.read_csv(io.BytesIO(results), sep="\t", date_parser=True)
    results.columns = [col.strip() for col in results.columns]
    results.index = results[results.columns[0]]
    results.pop(results.columns[0])
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.plot(results.iloc[0], marker="o")
    ax.plot(results.iloc[1], marker="o")
    ax.plot(results.iloc[2], marker="o")
    ax.plot(results.iloc[3], marker="o")
    ax.set_yscale('log')
#     ax.set_xscale('log')
    ax.legend(labels=results.index)
    plt.show()
    return fig, ax
```

```python
field_size = b"""
framework/field size 	(64, 64) 	(128, 128) 	(256, 256) 	(512, 512) 	(1024, 1024)
numpy 	0.695 	1.85 	9.56 	39.5 	148.0
JAX (CPU) 	0.336 	0.904 	2.94 	11.1 	45.0
JAX (GPU) 	0.193 	0.189 	0.199 	0.237 	0.613
JAX (TPU) 	0.059 	0.074 	0.119 	0.272 	0.842
"""
fig, ax = plot(field_size)
fig.savefig("results/field_size.png")
```
